```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [12, 9]
```

```python
frizzy = cv2.imread('./images/frizzy.png', cv2.IMREAD_GRAYSCALE)
froomer = cv2.imread(('./images/froomer.png'), cv2.IMREAD_GRAYSCALE)

plt.subplot(1, 2, 1)
plt.imshow(frizzy, cmap='gray')
plt.title('frizzy'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 2, 2)
plt.imshow(froomer, cmap='gray')
plt.title('froomer'), plt.xticks([]), plt.yticks([]);
```

```python
plt.imshow(cv2.Canny(frizzy, 100, 255) & cv2.Canny(froomer, 100, 255), cmap='gray');
plt.title('secret code'), plt.xticks([]), plt.yticks([]);
```
