## Add 2 Images

```python
dolphin = cv2.imread('./images/dolphin.png', cv2.IMREAD_GRAYSCALE)
bicycle = cv2.imread('./images/bicycle.png', cv2.IMREAD_GRAYSCALE)

summed = dolphin + bicycle
plt.imshow(summed, cmap='gray');
```

```python
average = (dolphin / 2) + (bicycle / 2)
plt.imshow(average, cmap='gray');
```

```python
average_alt = (dolphin + bicycle) / 2
plt.imshow(average_alt, cmap='gray');
```

## Multiply by a Scalar

```python
dolphin = cv2.imread('./images/dolphin.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(dolphin*1.5, cmap='gray');
```

## Blend 2 Images

```python
def blend(a, b, alpha):
    return alpha * a + (1 - alpha) * b

dolphin = cv2.imread('./images/dolphin.png', cv2.IMREAD_GRAYSCALE)
bicycle = cv2.imread('./images/bicycle.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(blend(dolphin, bicycle, 0.85), cmap='gray');
```

## Image Difference

```python
dolphin = cv2.imread('./images/dolphin.png', cv2.IMREAD_GRAYSCALE)
bicycle = cv2.imread('./images/bicycle.png', cv2.IMREAD_GRAYSCALE)

diff = (dolphin - bicycle)

plt.imshow(diff, cmap='gray');
```
