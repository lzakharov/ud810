# 2A-L3 Linearity and convolution

## Explore Edge Options

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

img = cv2.cvtColor(cv2.imread('./images/fall-leaves.png'), cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

### Apply Gaussian filter, specifying odge parameter

```python
from scipy.ndimage import gaussian_filter
```

```python
smoothed = gaussian_filter(img, sigma=0.8, mode='reflect')
plt.imshow(smoothed);
```

```python
smoothed = gaussian_filter(img, sigma=0.8, mode='constant')
plt.imshow(smoothed);
```

```python
smoothed = gaussian_filter(img, sigma=0.8, mode='nearest')
plt.imshow(smoothed);
```

```python
smoothed = gaussian_filter(img, sigma=0.8, mode='mirror')
plt.imshow(smoothed);
```

```python
smoothed = gaussian_filter(img, sigma=0.8, mode='wrap')
plt.imshow(smoothed);
```

## Unsharp Mask

```python
img = cv2.cvtColor(cv2.imread('./images/blood-cells.png'), cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

```python
unsharp_mask = -np.ones((3, 3)) / 9
unsharp_mask[1, 1] += 2

img_sharped = cv2.filter2D(img, -1, unsharp_mask)
plt.imshow(img_sharped);
```

## Median Filter

```python
img = cv2.cvtColor(cv2.imread('./images/moon.png'), cv2.IMREAD_GRAYSCALE)
plt.imshow(img);
```

### Add salt & pepper noise

```python
from skimage.util import random_noise
from skimage import img_as_ubyte

salted_and_peppered = img_as_ubyte(random_noise(img, mode='s&p'))
plt.imshow(salted_and_peppered);
```

### Apply a median filter

```python
median_filtered = cv2.medianBlur(salted_and_peppered, 3)
plt.imshow(median_filtered);
```
